```python
import sympy as sp

#Question A
x,y = sp.symbols('x y')
f = sp.exp(x) * sp.sin(y) + y**3
df_dx= sp.diff(f,x) #partial derivative with respect to x
df_dy=sp.diff(f,y) #partial derivative with respect to y
print(df_dx)
print(df_dy)
```

```python
#Question B
g=(x**2*y)+(x*y**2)
dg_dx=sp.diff(g,x) #partial derivative with respect to x
dg_dy=sp.diff(g,y) #partial derivative with respect to y
print(dg_dx)
print(dg_dy)
gvx=dg_dx.subs({x:1,y:-1})
gvy=dg_dy.subs({x:1,y:-1})
magnitude=sp.sqrt(gvx**2 + gvy**2)
print(magnitude)
```

```python
#Question C
h=sp.ln(x**2+y**2)
dh_dxx=sp.diff(sp.diff(h,x), x) #second partial derivative with respect to x
dh_dyy=sp.diff(sp.diff(h,y), y) #second partial derivative with respect to y
dh_dxy=sp.diff(sp.diff(h,x), y) #second partial derivative with respect to y of the partial derivative with respect to x
print(dh_dxx)
print(dh_dyy)
print(dh_dxy)
```

```python
#Question D
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
#Define the function using SymPy:
x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3
#Convert the SymPy expression to a numerical function:
j_func = sp.lambdify((x, y), j, 'numpy')
#Create a grid of x and y values and evaluate j on this grid:
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)
#Plot the contour:
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
